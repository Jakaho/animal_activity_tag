```python
!pip install memory_profiler
%load_ext memory_profiler

import numpy as np
import sys
sys.path.append('../')  # Adjust the path as necessary
import dataprocessing



import pandas as pd

filepath = '../datasets/pos_A.csv'
windows_ac = dataprocessing.import_and_downsample(filepath, False)  #
all_time_domain_signals = dataprocessing.process_window(windows_ac, None)
features_df = pd.DataFrame()
```

```python
window_features = []

for window in windows_ac:
    # Existing mean calculations
    mean_ax = window['ax'].mean()
    mean_ay = window['ay'].mean()
    mean_az = window['az'].mean()
    
    # New feature: standard deviation calculations
    std_ax = window['ax'].std()
    std_ay = window['ay'].std()
    std_az = window['az'].std()
    
    # Assuming the label is constant within a window
    label = window['label'].iloc[0]
    
    # Update the dictionary to include new features
    window_dict = {
        'mean_acc_x': mean_ax,
        'mean_acc_y': mean_ay,
        'mean_acc_z': mean_az,
        'std_acc_x': std_ax,
        'std_acc_y': std_ay,
        'std_acc_z': std_az,
        'label': label
    }
    
    # Append the updated dictionary to the list
    window_features.append(window_dict)

# Convert the list of dictionaries to a DataFrame
features_df = pd.DataFrame(window_features)
features_df.index.name = 'window_id'


```

```python
variance_acc_x = []
variance_acc_y = []
variance_acc_z = []
max_acc_x = []
min_acc_x = []
max_acc_y = []
min_acc_y = []
max_acc_z = []
min_acc_z = []

for window in windows_ac:
    var_ax = window['ax'].var()
    var_ay = window['ay'].var()
    var_az = window['az'].var()
    
    variance_acc_x.append(var_ax)
    variance_acc_y.append(var_ay)
    variance_acc_z.append(var_az)
    
    max_acc_x.append(window['ax'].max())
    min_acc_x.append(window['ax'].min())
    max_acc_y.append(window['ay'].max())
    min_acc_y.append(window['ay'].min())
    max_acc_z.append(window['az'].max())
    min_acc_z.append(window['az'].min())
    
features_df['variance_acc_x'] = variance_acc_x
features_df['variance_acc_y'] = variance_acc_y
features_df['variance_acc_z'] = variance_acc_z
features_df['max_acc_x'] = max_acc_x
features_df['min_acc_x'] = min_acc_x
features_df['max_acc_y'] = max_acc_y
features_df['min_acc_y'] = min_acc_y
features_df['max_acc_z'] = max_acc_z
features_df['min_acc_z'] = min_acc_z

```
